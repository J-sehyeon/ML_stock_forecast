## Driver 인스턴스 생성

```python
### Driver creation example
from neo4j import GraphDatabase
uri = "neo4j://127.0.0.1:7687"
driver = GraphDatabase.driver(
    uri, 
    auth=('neo4j', 'testpassword')
)

driver.verify_connectivity()
```

```python
# execute_query() method executes a Cypher query and returns the results
records, summary, keys = driver.execute_query(
    "RETURN COUNT {()} AS count"
)

# get the first record
first = records[0]

# print the count entry
print(first['count'])

driver.close()  # 위에 생성한 드라이버 종료
```

```python
# 드라이브 생성과 종료를 한번에 선언
with GraphDatabase.driver(uri, auth=('neo4j', 'testpassword')) as driver:
    result, summary, keys = driver.execute_query("RETURN COUNT {()} as count")
```
